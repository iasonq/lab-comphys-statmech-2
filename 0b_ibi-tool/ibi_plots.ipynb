```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.array(["Na-Na","Cl-Cl", "Na-Cl"]) #pair
```

```python
def loaddata(i):
    r, g    = np.loadtxt("./"+p[i]+"/output_g_"+p[i]+".txt", unpack=True)
    rs, gs  = np.loadtxt("./"+p[i]+"/output_g_"+p[i]+".txt_smt", unpack=True)
    rv, v   = np.loadtxt("./"+p[i]+"/pot-output_g_"+p[i]+".txt_0", unpack=True)
    rvs, vs = np.loadtxt("./"+p[i]+"/pot-output_g_"+p[i]+".txt_smt_0", unpack=True)
    return g, r, gs, rs, v, rv, vs, rvs
```

```python
fig, axs = plt.subplots(3, 2, figsize=(22.00, 10.00))
for i in range(0, 3):
    g, r, gs, rs, v, rv, vs, rvs = loaddata(i)

    axs[i, 0].plot(r, g, label="data")
    axs[i, 0].plot(rs, gs, 'tab:red', label="smoothed")
    axs[i, 0].legend(loc="upper right")
    axs[i, 0].set_title('RDF '+p[i])

    axs[i, 1].plot(rv, v, label="data")
    axs[i, 1].plot(rvs, vs, 'tab:red', label="smoothed")
    axs[i, 1].legend(loc="upper right")
    axs[i, 1].set_title('Potential '+p[i])
    axs[i, 1].set_xlim( (-2, 12) )
    axs[i, 1].set_ylim( (-5, 40) )
fig.show()
```
